```python
# Libreria necesaria para ejecutar comandos en SQL en Python

import sqlite3
```

```python
# Crear una coneccion a una Base de Datos
coneccion = sqlite3.connect('webshop')

# Crear un cursor que permite ejecutar comandos SQL
cursor = coneccion.cursor()
```

### Crear las tablas para la empresa WebShop

```python
cursor.execute(''' CREATE TABLE productos
                (id_prod INTEGER PRIMARY KEY AUTOINCREMENT,
                nombre CHAR(30) NOT NULL,
                precio FLOAT NOT NULL,
                stock INTEGER NOT NULL)
''')
```

```python
cursor.execute(''' CREATE TABLE clientes
                (rut_cliente INTEGER PRIMARY KEY AUTOINCREMENT,
                nombre CHAR(30) NOT NULL,
                calle CHAR(30) NOT NULL,
                comuna CHAR(30) NOT NULL)
''')
```

```python
cursor.execute(''' CREATE TABLE ventas
                (id_venta INTEGER PRIMARY KEY AUTOINCREMENT,
                rut_cliente INTEGER NOT NULL,
                fecha DATE NOT NULL,
                comuna CHAR(30) NOT NULL,
                estatus CHAR(50) NOT NULL,
                FOREIGN KEY (rut_cliente) REFERENCES clientes (rut_cliente))
''')
```
